Use the .dropna() method to drop NA values in data frame

```python
clean_df = df.dropna()
```

```python
clean_df.tail()
```

To access a particular column from a data frame we use the square bracket notation

```python
clean_df['Starting Median Salary']
```

To find the highest value we simply chain the .max() method.

```python
clean_df['Starting Median Salary'].max()
```

.idxmax() method will give us index for the row with the largest value.

```python
clean_df['Starting Median Salary'].idxmax()
```

To see name of the colum that corresponds to that particular row, we can use the .loc (location) property.

```python
clean_df['Undergraduate Major'].loc[43]
```

Selecting both a column ('Undergraduate Major') and a row at index 43, we can retrieve the value of a particular cell.

```python
clean_df['Undergraduate Major'][43]
```

If you don't specify a particular column you can use the .loc property to retrieve an entire row:

```python
clean_df.loc[43]
```
